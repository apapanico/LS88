##### 3. Offensive Eras in Baseball
Baseball is known to have offensive "eras".  The Dead-Ball era was an era of low offensive performance from about 1900-1920.  The Live-Ball era started around 1920 and it is believed this was due to a change in the manufacturing of the ball (See [ringer.com][1]).  The 1960s featured a dramatic tilt towards pitchers with the mound being lowered in 1969.  More recently we had the 1990s and early 2000s Steroid era.  Most recent is the 2010s decline and then resurgence (again, see [ringer.com][1] and [538.com][2]).  

To what extent to each of the classical statistics indicate the variability in performance across these eras?
